```python
import xml.etree.ElementTree as ET
import numpy as np
from PIL import Image

def create_mask_from_voc(xml_path):
    # Parse the XML file
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Get image dimensions
    width = int(root.find('.//size/width').text)
    height = int(root.find('.//size/height').text)

    # Create an empty black mask (0 = background)
    mask = np.zeros((height, width), dtype=np.uint8)

    # Iterate through each object and fill its bounding box
    for obj in root.findall('object'):
        name = obj.find('name').text

        # Get bounding box coordinates
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # Fill the rectangle area with 255 (white)
        # For multi-class, you can map specific names to unique integer IDs instead
        mask[ymin:ymax, xmin:xmax] = 255

    return Image.fromarray(mask)

# Generate and save the mask
mask_image = create_mask_from_voc('/content/drive/MyDrive/Thesis/anik_mahera_deep_back/run-2/objectdetectionphenotypingdataset/train/annotations/#9_translated_nalidixate_4.xml')
mask_image.save('mask_output.png')
```
